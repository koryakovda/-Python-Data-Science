### Задание 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
boston.keys()
```

```python
X = pd.DataFrame(boston.data,columns=boston.feature_names)

Y = boston.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(X,Y, test_size=0.30,random_state=42)
X_train.shape, X_test.shape, Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,Y_pred)
```

### Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=1000,max_depth = 12,random_state = 42)
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
r2_score(Y_test,Y_pred)
```

##### Вывод: RandomForestRegressor работает лучше

### Задание 4

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV,train_test_split
pd.options.display.max_columns = 100
```

```python
df = pd.read_csv('../lesson2/creditcard.csv') 
```

```python
df.value_counts(normalize = True)
```

```python
df.info()
```

```python
df.head(10)
```

```python
X = df.drop('Class', axis = 1)
y = df['Class']
```

```python
X_train,X_test,y_train,y_test = train_test_split (X,y,test_size=0.3, random_state=100, stratify = y)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15],'max_features': np.arange(3, 5),'max_depth': np.arange(4, 7)}]
clf = GridSearchCV(estimator = RandomForestClassifier(random_state=100),param_grid=parameters,scoring='roc_auc',cv=3)
clf.fit(X_train,y_train)
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
y_pred_proba = clf.predict_proba(X_test)
y_pred_proba_train = clf.predict_proba(X_train)
```

```python
y_pred
```

```python
y_pred_proba = y_pred_proba[:,1]
y_pred_proba_train = y_pred_proba_train[:,1]
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,y_pred_proba)
```

```python
roc_auc_score(y_train,y_pred_proba_train)
```

```python
target = pd.DataFrame(data.target, columns = ['Target'],dtype = np.int64)
target.head()
```

```python
X
```

```python
import seaborn as sns
X_corr = X.corr()
plt.figure(figsize = (20,15))
sns.set(font_scale = 1.4)
sns.heatmap(X_corr.abs(), annot = True, linewidth = 1, cmap = 'viridis')
plt.title('Correlation matrix')
plt.show()
```

```python
high_corr = []
for i in data_keys:
    if X_corr['Target'].abs()[i] > 0.5:
        high_corr.append(i)
high_corr
```

```python
X = X.drop('Target',axis=1)
```

```python
for i in high_corr:
    X[f'{i}_2'] = X[i]**2
X
```
